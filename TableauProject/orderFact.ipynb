```python
import pandas as pd
orders = pd.read_csv('orders.csv',sep=';')
orders['Discount'] = pd.to_numeric(orders['Discount'].str.replace(',', '.'),errors='coerce',downcast='float')
orders['Profit'] = pd.to_numeric(orders['Profit'].str.replace(',', '.'))
orders['Sales'] = pd.to_numeric(orders['Sales'].str.replace(',', '.'))

orders['Discount'].head()
# orders['Profit'].head()
# orders['Sales'].head()
orders.head()

orders.to_csv('updated_orders.csv', index=False,sep=',')
```

```python
orders_excel = pd.read_csv('updated_orders.csv',sep=',')
orders_excel.head()
orders_excel.to_excel('updated_orders.xlsx', index=False)
```
